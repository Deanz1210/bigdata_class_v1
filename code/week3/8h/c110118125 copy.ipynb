```python
import requests
from bs4 import BeautifulSoup
import re
from fake_useragent import UserAgent
from datetime import datetime
```

```python
user_agent = UserAgent()
user_agent.random
```

# 網頁

```python
category_url = 'https://gnn.gamer.com.tw/index.php?k=5'

req = requests.get(category_url, headers={ 'user-agent': user_agent.random }, timeout=5)
print(req.status_code)

print(req.text)
```

# 解析 HTML 內容

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

# 找到具有特定 class 名稱的 li 元素

```python
news_items = soup.find_all('div', class_='GN-lbox2B')
news_list = []
print(news_items)
```

# 找出最新電競賽事新聞的每個標題與連結

```python



# 遍歷每個新聞項目
for item in news_items:
    # 找標題 h3
    title_tag = item.find('h1', class_='GN-lbox2D')
    # 找超連結 a
    link_tag = item.find('a', href=True)

    if title_tag and link_tag:
        title = title_tag.text.strip()
        link = link_tag['href']

        # 處理相對連結 (補上 Yahoo 網址)
        if link.startswith('/'):
            link = f"https://gnn.gamer.com.tw{link}"

        # 加入新聞列表
        news_list.append({'標題': title, '連結': link})

# 顯示新聞列表
for news in news_list:
    print(f"標題: {news['標題']}")
    print(f"連結: {news['連結']}")
    print()
```

# 其中一條新聞內容網頁

```python
category_url = 'https://gnn.gamer.com.tw/detail.php?sn=282736'

req = requests.get(category_url, headers={ 'user-agent': user_agent.random }, timeout=5)
```

```python
soup = BeautifulSoup(req.text, 'html.parser')
```

# 取得新聞時間標籤

```python
time_element = soup.find('span', class_='GN-lbox3C')

if time_element:
            text_content = time_element.text.strip()
            parts = text_content.split()  

            try:
                news_time = datetime.strptime(parts[-2] + " " + parts[-1], "%Y-%m-%d %H:%M:%S")
                dtime = news_time.strftime("%Y/%m/%d %H:%M")  
                news_date = news_time.strftime("%Y-%m-%d")  
            except Exception as e:
                print(f"日期格式錯誤: {e}")
                dtime = "未知"
                news_date = "未知"
else:
    dtime = "未知"
    news_date = "未知"
    
print(f"發布日期: {news_date}, 發布時間: {dtime}")
```

# 找到主要圖片連結

```python
image_url = None

# **方法 1: 嘗試直接抓取 <figure> 內的 <img>**
figure_tag = soup.find("figure", class_="caas-figure")
if figure_tag:
    img_tag = figure_tag.find("img")
    if img_tag:
        image_url = img_tag.get("src") or img_tag.get("data-src")  # 優先抓 src，若無則抓 data-src


# **Yahoo 可能返回錯誤的圖片網址 (包含兩個 URL)，需要修正**
if image_url and "https://s.yimg.com/ny/api/res/" in image_url:
    match = re.search(r'(https://s\.yimg\.com/os/creatr-uploaded-images/[\w./-]+)', image_url)
    if match:
        image_url = match.group(0)

# **輸出最終圖片網址**
if image_url:
    print("新聞圖片網址:", image_url)
else:
    print("未找到圖片")
```

# 抓取新聞內容 並一段段列出

```python
paragraphs = [p.text for p in soup.select(".caas-body p")]

for i, text in enumerate(paragraphs, 1):
    print(f"{i}. {text}")
```

# 移除相關新聞內容段落

```python

paragraphs = [p.text.strip() for p in soup.select(".caas-body p")]
filtered_news = [item for item in paragraphs if not item.startswith("相關新聞")]

for i, text in enumerate(filtered_news, 1):
    print(f"{i}. {text}")
```
