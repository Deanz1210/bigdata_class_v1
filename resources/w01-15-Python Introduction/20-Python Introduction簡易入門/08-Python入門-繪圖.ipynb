## matplotlib繪圖

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 3*np.pi, 50)
plt.plot(x, np.sin(x))
plt.title('A simple chart');
```

## pandas繪圖功能

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot()
```

# 複雜一些: 圖形排排站

```python
# plot 10 images
plt.figure(figsize=(20,10))
plt.suptitle('Charts', fontsize=28)

for i in range(8):
    plt.subplot(2, 4, i+1) # 這是甚麼意思? #rows, #columns, ith chart (start with 1, not 0)
    
    #plt.grid(False)
    #plt.imshow(  ) # you may open an photo to show
    
    x = np.linspace(0, 3*np.pi, 50)
    plt.plot(x, np.sin(x))
    plt.title('A simple chart');
    
plt.show()
```
